**Fig. 1** Percentage of organic and inorganic phosphorus removed from soil samples using various methods. Unhatched bars indicate sample data collected from a grazing pasture, while hatched bars indicate sample data collected from a wheat field. Residual P is plotted against the right-hand y-axis and is separated from previous methods with a dashed line. 

Resin and NaHCO3 extracted P is about 1% higher in pasture soils versus wheat field soils. HCl extracted P shows the greatest difference between sites, with wheat field soils being about 4% higher than pasture soils. Residual P is about 1% higher in pasture soils than wheat field soils. Other methods showed less than 1% difference of extracted P between pasture and wheat field soils. Resin extracted P represents mobile/free floating P and NaHCO3 extracted P represents P that is loosely adsorbed to soil surfaces. These types of P may be higher in the pasture soils due to less plant uptake occurring. In the wheat fields, we may expect the crops to be uptaking readily accessible P. This increased uptake might lead to a decreased proportion of Resin and NaHCO3 extractable P in wheat field soils. HCl extracted P represents P that was occluded in the crystal structure of the mineral apatite. As previously mentioned, if plant uptake is decreasing the relative proportion of readily accessible P, then the relative amount of occluded P would be higher. This may explain the higher proportion of HCl extracted P in wheat field soils, since plants could not uptake this type of P. Regarding proportions of organic versus inorganic P extraction, NaHCO3 shows the greatest difference between wheat field and pasture soils. NaHCO3 extracted organic P is less than 0.5 percent lower in the wheat field than the pasture. However, extracted inorganic P is more than one percent lower in the wheat field. This lower proportion of inorganic P may be explained by agricultural land practices, such as soil tilling. Soil tilling would break up sediments, reducing the amount of soil adsorbed P and increasing the probability of inorganic P being uptaken by the crops. This would reduce the relative amount of NaHCO3 extracted inorganic P compared to organic P in the wheat field.

In reality, the soils could experience various degrees of compaction according to topography, plants present, human/animal activity, and more. # Code for Generating Figures

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
```

```python
#Import data
df = pd.read_csv('HW3_extract_results.csv')
df
```

```python
#Sum P for both Pasture and Wheat
P_p_sum = df['Pasture'].sum()
P_w_sum = df['Wheat'].sum()
print(P_p_sum)
print(P_w_sum)
```

```python
#Create % columns
df['Pasture_%'] = df['Pasture'].apply(lambda x: 100*x/P_p_sum)
df['Wheat_%'] = df['Wheat'].apply(lambda x: 100*x/P_w_sum)
df
```

```python
# Make the pools into columns so bar graphs are easy to make
df = df.pivot(columns='Pool', index='Method', values=['Pasture_%','Wheat_%'])
df
```

```python
# GRAPHING
#################

# Colors for bars (inorganic, organic, residual)
bar_colors = ('gray', 'lightgray', '#3a3a3a')

# Reorder columns so the bars in the figure are ordered correctly
# Also make seperate pasture/wheat dataframes
pool_order = ['Inorganic','Organic','Residual']
method_order = ['Resin','NaHCO3','CHCl-released NaHCO3','NaOH','Sonicate-NaOH','HCl','H2SO4/H2O2']
P_df = df['Pasture_%'].loc[method_order,pool_order]
W_df = df['Wheat_%'].loc[method_order,pool_order]

# Add y label now so the current axis doesn't switch to the secondary y
fig,ax = plt.subplots()
plt.ylabel('% P removed')

# Plot pasture and wheat dataframes
P_df.plot.bar(ax=ax,stacked=True,position=1,width=0.2, color=bar_colors, edgecolor='black',secondary_y='Residual')
W_df.plot.bar(ax = ax, stacked=True,position=0, width=.2, hatch='//', color=bar_colors, edgecolor='black',secondary_y='Residual')

# Add dashed vertical line
plt.axvline(x=5.5, color='black', linestyle='--')

# Create custom legend elements and add legend
legend_el = [Patch(facecolor=bar_colors[0], edgecolor='black', label='Inorganic'),
            Patch(facecolor=bar_colors[1], edgecolor='black', label='Organic'),
            Patch(facecolor=bar_colors[2], edgecolor='black', label='Residual'),
            Patch(facecolor='white', edgecolor='black', label='Pasture'),
            Patch(facecolor='white', edgecolor='black', label='Wheat', hatch='//')]
ax.legend(handles=legend_el, ncol=2)

# Right axis y-label
plt.ylabel('% P removed (residual)')

# Split long tick label into two lines and format tick labels
x_labels = [item.get_text() for item in ax.get_xticklabels()]
x_labels[2] = 'CHCl-released\nNaHCO3'
ax.set_xticklabels(x_labels, rotation=45, fontsize=7)

plt.show()
```
